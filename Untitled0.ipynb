```python

!pip install numpy matplotlib torch

import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
import time
from IPython.display import clear_output

class PlasmaSimulator:
    def __init__(self):
        self.target_ip = 1.5
        self.target_beta = 2.4
        self.reset()

    def reset(self):
        self.ip = 1.2 + np.random.random() * 0.6
        self.beta = 1.8 + np.random.random() * 1.2
        self.ne = 0.5 + np.random.random() * 0.8
        self.te = 3.0 + np.random.random() * 4.0
        self.tearing = 0.05 + np.random.random() * 0.1
        self.z = (np.random.random() - 0.5) * 0.1
        self.step_count = 0
        return self.get_state()

    def get_state(self):
        return np.array([self.ip, self.beta, self.ne, self.te, self.tearing, self.z])

    def step(self, action):
        if action == 0:
            self.ip += 0.05
        elif action == 1:
            self.ip -= 0.05
        elif action == 2:
            self.beta += 0.03
        elif action == 3:
            self.beta -= 0.03
        elif action == 4:
            self.te += 0.2

        ip_error = abs(self.ip - self.target_ip) / self.target_ip
        beta_error = abs(self.beta - self.target_beta) / self.target_beta
        inst_drive = ip_error + beta_error + abs(self.z) * 0.5

        self.tearing += inst_drive * 0.02
        self.tearing *= 0.98

        self.ip = np.clip(self.ip, 0.6, 2.8)
        self.beta = np.clip(self.beta, 0.5, 4.5)
        self.ne = np.clip(self.ne, 0.2, 2.0)
        self.te = np.clip(self.te, 1.0, 15.0)
        self.z = np.clip(self.z, -0.2, 0.2)

        ip_err = abs(self.ip - self.target_ip) / self.target_ip
        beta_err = abs(self.beta - self.target_beta) / self.target_beta
        inst_penalty = self.tearing * 2
        pos_penalty = abs(self.z) * 5
        reward = np.exp(-3 * (ip_err + beta_err)) * 10 - inst_penalty - pos_penalty

        done = self.tearing > 0.85 or abs(self.z) > 0.15

        self.step_count += 1
        return self.get_state(), reward, done

class ActorCritic(nn.Module):
    def __init__(self, state_dim=6, action_dim=5):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(state_dim, 32),
            nn.ReLU(),
            nn.Linear(32, 32),
            nn.ReLU()
        )
        self.actor = nn.Linear(32, action_dim)
        self.critic = nn.Linear(32, 1)

    def forward(self, state):
        x = self.net(state)
        probs = torch.softmax(self.actor(x), dim=-1)
        value = self.critic(x)
        return probs, value

class PPOAgent:
    def __init__(self):
        self.net = ActorCritic()
        self.optimizer = optim.Adam(self.net.parameters(), lr=1e-3)
        self.gamma = 0.99
        self.epsilon = 0.2
        self.buffer = []

    def act(self, state):
        state = torch.FloatTensor(state).unsqueeze(0)
        with torch.no_grad():
            probs, _ = self.net(state)
        action = torch.multinomial(probs, 1).item()
        return action

    def remember(self, state, action, reward, next_state, done):
        self.buffer.append((state, action, reward, next_state, done))

    def train(self):
        if len(self.buffer) < 32:
            return

        states, actions, rewards, next_states, dones = zip(*self.buffer[-500:])
        states = torch.FloatTensor(np.array(states))
        actions = torch.LongTensor(actions)
        rewards = torch.FloatTensor(rewards)
        next_states = torch.FloatTensor(np.array(next_states))
        dones = torch.FloatTensor(dones)

        with torch.no_grad():
            _, next_values = self.net(next_states)
            targets = rewards + self.gamma * next_values.squeeze() * (1 - dones)

        probs, values = self.net(states)
        probs = probs.gather(1, actions.unsqueeze(1)).squeeze()

        advantages = targets - values.squeeze()

        old_probs = probs.detach()
        ratio = probs / (old_probs + 1e-8)
        clipped_ratio = torch.clamp(ratio, 1 - self.epsilon, 1 + self.epsilon)
        loss = -torch.min(ratio * advantages, clipped_ratio * advantages).mean()
        loss += (advantages ** 2).mean() * 0.5

        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()

        self.buffer = []

print("ОБУЧЕНИЕ PPO НА СИМУЛЯТОРЕ ПЛАЗМЫ")
print("=" * 60)

env = PlasmaSimulator()
agent = PPOAgent()
rewards_history = []

for episode in range(200):
    state = env.reset()
    total_reward = 0
    steps = 0

    for step in range(100):
        action = agent.act(state)
        next_state, reward, done = env.step(action)

        agent.remember(state, action, reward, next_state, done)
        total_reward += reward
        state = next_state
        steps += 1

        if done:
            break

    agent.train()
    rewards_history.append(total_reward)

    if episode % 10 == 0:
        clear_output(wait=True)
        avg_reward = np.mean(rewards_history[-10:])

        plt.figure(figsize=(12, 4))

        plt.subplot(1, 2, 1)
        plt.plot(rewards_history)
        plt.xlabel('Эпизод')
        plt.ylabel('Награда')
        plt.title('Кривая обучения')
        plt.grid(True)

        plt.subplot(1, 2, 2)
        plt.plot(rewards_history[-50:])
        plt.xlabel('Последние 50 эпизодов')
        plt.ylabel('Награда')
        plt.title(f'Средняя награда: {avg_reward:.2f}')
        plt.grid(True)

        plt.tight_layout()
        plt.show()

        print(f"Эпизод {episode:3d} | Награда: {total_reward:6.2f} | Длит: {steps:2d} | Средняя: {avg_reward:6.2f}")

print("=" * 60)
print(" ОБУЧЕНИЕ ЗАВЕРШЕНО!")
print(f"\nИтоговая средняя награда (последние 20 эпизодов): {np.mean(rewards_history[-20:]):.2f}")
print(f"Максимальная награда: {np.max(rewards_history):.2f}")
print(f"Минимальная награда: {np.min(rewards_history):.2f}")
```
